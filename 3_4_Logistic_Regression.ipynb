```python
import numpy as np, matplotlib.pyplot as plt, seaborn as sns
from sklearn import metrics
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

**3.4 Logistic Regression** - A model of the log-odds of an event with any number of variables greater than 0. This is done by estimating the parameters of a logistic model. This estimation is carried out by using Maximum-Likelihood Estimation (MLE). It's important to note that this can simplify problems because it assumes that at some point, the probability is 100%.

```python
dataset = load_digits()
for index, (image, label) in enumerate(zip(dataset.data[0:5], dataset.target[0:5])):
 plt.subplot(1, 5, index+1)
 plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
 plt.title('Training: %i\n' % label, fontsize = 15)
 plt.tight_layout()
```

```python
x_train, x_test, y_train, y_test = train_test_split(dataset.data, dataset.target, test_size=0.25, random_state=0)
logReg = LogisticRegression(max_iter=1000)
logReg.fit(x_train, y_train)
```

```python
logReg.predict(x_test[0:10])
```

```python
predictions = logReg.predict(x_test)
score = logReg.score(x_test, y_test)
print(score)
```

```python
conMatrix = metrics.confusion_matrix(y_test, predictions)
plt.figure(figsize=(9,9))
sns.heatmap(conMatrix, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0:.3f}'.format(score)
plt.title(all_sample_title, size = 15);
```
